```python
import nltk
from nltk.probability import FreqDist
from nltk.corpus import stopwords
import pandas as pd
```

```python
stop_words=set(stopwords.words('english'))
print(stop_words)
```

```python
#read the bryant corpus
words_bryant = nltk.Text(nltk.corpus.gutenberg.words('bryant-stories.txt'))
print(len(words_bryant))
print(words_bryant)
```

```python
#read the austen-emma corpus
words_emma = nltk.Text(nltk.corpus.gutenberg.words('austen-emma.txt'))
print(len(words_emma))
print(words_emma)
```

```python
#for words_bryant

#convert to small letters
words_bryant=[word.lower() for word in words_bryant if word.isalpha()]
print(len(words_bryant))

#remove stoping words
words_bryant=[word.lower() for word in words_bryant if word not in stop_words]
print(len(words_bryant))
```

```python
#for words_emma

#convert to small letters
words_emma=[word.lower() for word in words_emma if word.isalpha()]
print(len(words_emma))

#remove stoping words
words_emma=[word.lower() for word in words_emma if word not in stop_words]
print(len(words_emma))
```

```python
ttr_bryant=len(set(words_bryant))/len(words_bryant)
ttr_emma=len(set(words_emma))/len(words_emma)

print("TTR of bryant = ",ttr_bryant)
print("TTR of emma = ",ttr_emma)
```

```python
sr_bryant={'token':len(words_bryant),'Vocabulary':len(set(words_bryant)),'TTR':ttr_bryant}
sr_emma={'token':len(words_emma),'Vocabulary':len(set(words_emma)),'TTR':ttr_emma}
```

```python
print("For Bryant Stories: ",sr_bryant)
print("For Jane Austen emma: ",sr_emma)
```
